```python
def fun(cat):
    if cat=='Iris-setosa':
        return 0.0
    elif cat=='Iris-versicolor':
        return 1.0 
    else:
        return 2.0
  
# passing function to apply and storing returned series in new
df['Type']=df['Species'].apply(fun)
```

```python
df.drop(['Species'],inplace=True,axis=1)
df
```

```python
#df = df.astype('float64')
df.info()
```

```python
dataset=np.array(df)
X=dataset
y=dataset[100]
print("X= ",X)
print("y: ",y)
k=3
#calculate ED
ds=[]
for i in range(len(X)-1):
    d=np.subtract(X[i],y)
    t_s=np.square(d)
    rr=np.sum(t_s)
    ds=np.append(ds,[rr])
print('Distances= ',ds)
asd=np.argsort(ds)
print(asd)
sd=[]
for i in range(k):
    sd.append(X[asd[i],-1])
print("ASD: ",sd)
wx=sd.count(2.0)
wy=sd.count(1.0)
wz=sd.count(0.0)
if (wx>wy and wx>wz):
    print('Class label is 2.0')
elif(wy> wz):
    print('Class label is 1.0')
else:
    print('Class label is 0.0')
```
